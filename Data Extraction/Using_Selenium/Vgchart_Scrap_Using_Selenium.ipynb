```python
class Vgchart_Extract():
    from selenium.webdriver.common.by import By
    from selenium import webdriver
    import pandas as pd
    import time
    import random
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    
    def Header_Extraction(self):
        url='https://www.vgchartz.com/games/games.php?page=1&results=200&order=Sales&ownership=Both&direction=DESC&showtotalsales=1&shownasales=1&showpalsales=1&showjapansales=1&showothersales=1&showpublisher=1&showdeveloper=0&showreleasedate=1&showlastupdate=1&showvgchartzscore=1&showcriticscore=1&showuserscore=1&showshipped=1'
        driver=webdriver.Chrome('chromedriver',options=options)
        driver.get(url)
        var=[]
        for x in driver.find_elements(By.XPATH,'//tbody/tr[3]/th')[1:]:
                        var.append(x.text.strip())
        return var
    
    def Complete_Dataframe(self,Total_page=1,Total_record_per_page=10):
                    import pandas as pd
                    import random
                    try:
                        final_df=pd.DataFrame(columns=header_list)
                        for page in range(1,Total_page+1):
                                url1=f'https://www.vgchartz.com/games/games.php?page={page}&results=200&order=Sales&ownership=Both&direction=DESC&showtotalsales=1&shownasales=1&showpalsales=1&showjapansales=1&showothersales=1&showpublisher=1&showdeveloper=0&showreleasedate=1&showlastupdate=1&showvgchartzscore=1&showcriticscore=1&showuserscore=1&showshipped=1'
                                driver = webdriver.Chrome('chromedriver',options=options)
                                driver.get(url1)
                                final=[]
                                for i in range(Total_record_per_page):
                                            val=[]
                                            for x in driver.find_elements(By.XPATH,f"(//tr)[{28+i}]/td")[2:]:
                                                    if x.text.strip()!='':
                                                            val.append(x.text.strip())
                                                    else:
                                                        val.append(x.find_element(By.XPATH,"//td/img").get_attribute("alt"))
                                            intial_dict=dict(zip(header_list,val))
                                            final.append(intial_dict)
                                #genre_list=[]
                                second_dict=[]
                                for link in driver.find_elements(By.XPATH,"//tbody//tr//td/a[1]")[0:Total_record_per_page]:
                                            driver=webdriver.Chrome('chromedriver',options=options)
                                            driver.get(link.get_attribute('href'))
                                            
                                            tamp_tag=None
                                            if driver.find_elements(By.TAG_NAME,"h2")[1].text =="Genre":
                                                    tamp_tag=driver.find_element(By.XPATH,"(//h2[normalize-space()='Genre'])/following-sibling::p").text
                                    
                                            second_dict.append(tamp_tag)
                                            #genre_list.append(second_dict)
                                            
                                df=pd.DataFrame(final)
                                df['Genre']=second_dict
                                final_df=final_df.append(df,ignore_index=True)
                                time.sleep(random.randint(1,3))
                                driver.close()
                        return final_df
                    except NameError:
                        print("Error : Please Provide header_list")
```

```python
Vgchart_obj=Vgchart_Extract()
```
